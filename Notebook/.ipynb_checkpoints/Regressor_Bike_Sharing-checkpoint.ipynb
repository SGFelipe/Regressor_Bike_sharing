### 2.Prepare os dados utilizando as técnicas ensinadas na disciplina, quando pertinente.

```python
x = bike.drop('cnt', axis=1)
y = bike['cnt']
```

```python
#cores = y.astype('cnt').cat.codes
axes = pd.plotting.scatter_matrix(x, c=y, figsize=[15,15], marker='D')

for ax in axes.flatten():
    ax.xaxis.label.set_rotation(90)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')
```

```python
x = x.drop(['atemp','dteday'], axis=1).values # removendo variavel atemp por ser semelhante a temp
y = y.values
```

```python
x
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)
```

```python
print('X_train.shape', X_train.shape)
print('y_train.shape', y_train.shape)
print('X_test.shape', X_test.shape)
print('y_test.shape', y_test.shape)
```

```python
def build_regressors():
    regressors = []
    regressors.append(
        ('lr',
        LinearRegression(),
        {'fit_intercept': [True, False]}
        )
    )
    
    regressors.append(
        ('svr',
        SVR(),
        {'C': [0.1, 1, 10],
        'gamma': [0.00001, 0.0001, 0.001, 0.01, 1],
        'epsilon': [0.0001, 0.005, 0.001, 1, 5, 10]}
        )
    )
    
    regressors.append(
        ('rf',
        RandomForestRegressor(),
        {'n_estimators': [10, 50, 100]}
        )
    )
    
    regressors.append(
        ('dt',
        DecisionTreeRegressor(),
        {'max_depth': [2, 4, 6, 8, 10, 12]}
        )
    )
    
    return regressors
```

```python
train_results = []
test_results = []
names =[]

scoring = 'neg_root_mean_squared_error'

regressors = build_regressors()
kfold = KFold(n_splits=3, shuffle = True, random_state= 42)

for name, regressor, params in regressors:
    print(f'>>> Regressor:{name}')
    #Treinamento
    gs = GridSearchCV(regressor, param_grid=params, scoring = scoring, verbose = 1)
    gs.fit(X_train, y_train)
    cv_results = cross_val_score(gs, X_train, y_train, cv= kfold, scoring=scoring) #analise
    names.append(name)
    train_results.append(abs(cv_results))
    #teste
    y_pred = gs.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred)**0.5
    test_results.append(rmse)
    print('MAE', mean_absolute_error(y_test, y_pred))
    print('MSE', mean_squared_error(y_test, y_pred))
    print('RMSE', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
fig, ax = plt.subplots()
fig.suptitle('Desempenho em treinamento')
_=plt.boxplot(train_results)
ax.set_xticklabels(names)
ax.set_ylabel('RMSE')
ax.set_xlabel('Regressores')
```

```python
fig, ax = plt.subplots()
fig.suptitle('Desempenho em test')
_=plt.bar(names, test_results)
ax.set_ylabel('RMSE')
ax.set_xlabel('Regressores')
```

#### Melhor resultado obtido com Random Forest com as 3 metricas melhores do que os demais e agilidade no processamento.
